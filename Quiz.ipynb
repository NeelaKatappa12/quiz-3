```python
#@title Airbnb Host Analysis: Single vs. Multiple Listings

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots display correctly in Colab
%matplotlib inline

# =============================================================================
# Step 1: Provide the File Path Directly
# =============================================================================
# Paste the full file path to your Airbnb CSV file when prompted.
file_path = input('/content/listings_detailed.csv')

# =============================================================================
# Step 2: Load and Preview the Data
# =============================================================================
df = pd.read_csv(file_path)
print("Data Preview:")
display(df.head())

# =============================================================================
# Step 3: Determine the Host Listings Count Column
# =============================================================================
# Check for the most relevant column that indicates how many listings a host manages.
if 'host_listings_count' in df.columns:
    host_listings_col = 'host_listings_count'
elif 'calculated_host_listings_count' in df.columns:
    host_listings_col = 'calculated_host_listings_count'
else:
    raise ValueError("No host listings count column found in the dataset.")

print(f"Using '{host_listings_col}' as the listings count column.")

# =============================================================================
# Step 4: Create a Unique Hosts DataFrame and Analyze Distribution
# =============================================================================
# Since the dataset may contain multiple rows per host, create a unique host-level DataFrame.
df_hosts = df[['host_id', host_listings_col]].drop_duplicates()

# Distribution: How many hosts manage X listings?
listings_distribution = df_hosts[host_listings_col].value_counts().sort_index()
print("\nDistribution of Listings per Host:")
print(listings_distribution)

# Visualization: Bar Plot of the Distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=listings_distribution.index, y=listings_distribution.values, palette="viridis")
plt.xlabel("Number of Listings per Host", fontsize=14)
plt.ylabel("Number of Hosts", fontsize=14)
plt.title("Distribution of Listings per Host", fontsize=16)
plt.tight_layout()
plt.show()

# =============================================================================
# Step 5: Summarize Key Metrics and Visualize Proportions
# =============================================================================
total_hosts = len(df_hosts)
single_listing_hosts = len(df_hosts[df_hosts[host_listings_col] == 1])
multiple_listing_hosts = len(df_hosts[df_hosts[host_listings_col] > 1])

print(f"\nTotal Unique Hosts: {total_hosts}")
print(f"Hosts with a Single Listing: {single_listing_hosts} ({(single_listing_hosts/total_hosts)*100:.2f}%)")
print(f"Hosts with Multiple Listings: {multiple_listing_hosts} ({(multiple_listing_hosts/total_hosts)*100:.2f}%)")

# Visualization: Pie Chart of Single vs. Multiple Listing Hosts
labels = ['Single Listing', 'Multiple Listings']
sizes = [single_listing_hosts, multiple_listing_hosts]
colors = ['skyblue', 'lightgreen']
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Single vs. Multiple Listing Hosts", fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()

# =============================================================================
# Step 6: Managerial Insights
# =============================================================================
insights = "\n--- Managerial Insights ---\n"

# Insight 1: Host Concentration
multiple_pct = (multiple_listing_hosts / total_hosts) * 100
if multiple_pct > 20:
    insights += (
        f"1. A significant proportion of hosts ({multiple_pct:.2f}%) manage multiple listings. "
        "This suggests that professional property managers or speculative investors may be active on the platform, "
        "which could have implications for local housing markets.\n"
    )
else:
    insights += (
        f"1. The majority of hosts ({100 - multiple_pct:.2f}%) are single-listing owners, "
        "indicating that Airbnb remains predominantly a platform for private individuals.\n"
    )

# Insight 2: Average and Median Listings per Host
avg_listings = df_hosts[host_listings_col].mean()
median_listings = df_hosts[host_listings_col].median()
insights += f"2. The average number of listings per host is {avg_listings:.2f}, while the median is {median_listings}.\n"

# Insight 3: Strategic Considerations
insights += (
    "3. Understanding host concentration is critical for regulatory oversight and market analysis. "
    "Platforms with high proportions of multi-listing hosts may face different challenges compared to those dominated by private individuals.\n"
)

print(insights)
```
